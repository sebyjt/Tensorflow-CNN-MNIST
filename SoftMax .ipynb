```python
import tensorflow as tf
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist=input_data.read_data_sets("MNIST/",one_hot=True)
```

```python
mnist.train.num_examples
```

```python
import matplotlib.pyplot as plt
```

```python
single_image=mnist.train.images[0].reshape(28,28)
```

```python
plt.imshow(single_image,cmap='gist_gray')
```

```python
x=tf.placeholder(tf.float32,shape=[None,784])
```

```python
W=tf.Variable(tf.zeros([784,10]))
```

```python
B=tf.Variable(tf.zeros(10))
```

```python
y=tf.matmul(x,W)+B
```

```python
y_true=tf.placeholder(tf.float32,[None,10])
```

```python
cross_entropy=tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=y_true,logits=y))
```

```python
optimizer=tf.train.GradientDescentOptimizer(learning_rate=.5)
train=optimizer.minimize(cross_entropy)
```

```python
init=tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    for steps in range(1000):
        x_batch,y_batch=mnist.train.next_batch(100)
        sess.run(train,feed_dict={x:x_batch,y_true:y_batch})
        correct_prediction=tf.equal(tf.arg_max(y_true,1),tf.arg_max(y,1))
        acc=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
        print(sess.run(acc,feed_dict={x:mnist.test.images,y_true:mnist.test.labels}))
```
